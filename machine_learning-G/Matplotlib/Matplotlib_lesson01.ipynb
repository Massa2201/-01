```
# matplotlibの宣言をする
import matplotlib.pyplot as plt
import numpy as np
```

```
# 配列を作成し，折れ線を出力
x = np.array([1, 2, 3, 4, 5])
y = np.array([100, 200, 300, 400, 500])
# 描画
plt.plot(x, y)
# グラフ表示
plt.show()
```

```
x = np.arange(5)
y = np.arange(start=100, stop=600, step=100)
print("y=", y)
plt.plot(x, y)
plt.show()
```

```
volume = 30
x = np.arange(volume)
y = [x_i + np.random.randn(1) for x_i in x]
a, b = np.polyfit(x, y, 1)
_ = plt.plot(x, y, 'o', np.arange(volume), a * np.arange(volume) + b, '-')
```

```
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x))
plt.show()
```

```
plt.plot(x, np.sin(x))
plt.title("title01")
plt.xlabel("x-line")
plt.ylabel("y-line")
plt.grid(False)
plt.show()
```

```
x = np.linspace(0, 2 * np.pi)
plt.plot(x, np.sin(x))

plt.title("title")
positions = [0, np.pi / 2, np.pi, np.pi * 3 / 2, np.pi * 2]
labels = ["0", "90", "180", "270", "360"]
plt.xticks(positions, labels)
plt.show()
```

```
# 散布図
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(100)
y = np.random.rand(100)

plt.scatter(x, y)
plt.show()
```

```
plt.scatter(x, y, s=600, c="pink", alpha=0.5, linewidths="2", edgecolors="red")
plt.show()
```

```
plt.scatter(x, y, s=300, c=y, cmap="Greens")
plt.title("aaa")
plt.grid(True)
```

```
import matplotlib.pyplot as plt
plt.subplot(211)
plt.scatter(x, y, s=600, c="pink", alpha=0.5, linewidths="2", edgecolors="red")
plt.subplot(212)
plt.scatter(x, y, s=300, c=y, cmap="Greens")
plt.show()
```

```
import matplotlib.pyplot as plt
plt.subplot(121)
plt.scatter(x, y, s=600, c="pink", alpha=0.5, linewidths="2", edgecolors="blue")
plt.subplot(122)
plt.scatter(x, y, s=300, c=y, cmap="Greens")
plt.show()
```

```
plt.subplot(221)
plt.scatter(x, y, s=600, c="white", alpha=0.5, linewidths="2", edgecolors="yellow")
plt.subplot(222)
plt.scatter(x, y, s=300, c=y, cmap="Greens")
plt.subplot(223)
plt.scatter(x, y, s=600, c="pink", alpha=0.5, linewidths="2", edgecolors="red")
plt.subplot(224)
plt.scatter(x, y, s=300, c=y, cmap="Blues")
plt.show()
```

```
from mpl_toolkits.mplot3d import Axes3D

np.random.seed(0)
random_x = np.random.randn(100)
random_y = np.random.randn(100)
random_z = np.random.randn(100)

fig = plt.figure(figsize=(8, 8))

ax = fig.add_subplot(1, 1, 1, projection="3d")
x = np.ravel(random_x)
y = np.ravel(random_y)
z = np.ravel(random_z)

ax.scatter3D(x, y, z)
plt.show()
```

```
np.random.seed(0)

random_x = np.random.randn(100)
random_y = np.random.randn(100)
random_z = np.random.randn(100)

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(1, 1, 1, projection="3d")
x = np.ravel(random_x)
y = np.ravel(random_y)
z = np.ravel(random_z)
c = "r"
m = "^"
s = 300,

ax.scatter3D(x, y, z, s=s, c=c, marker=m)
plt.show()
```
